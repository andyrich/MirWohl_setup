```python
from scipy.interpolate import griddata
def interp(lay, array = None):
    
    if array is None:
        array = ml.dis.botm[lay,:,:]
    
    i,j = np.indices((ml.dis.nrow, ml.dis.ncol))
    
    
    mask = array >1e5
    # mask = ml.dis.botm[lay,:,:]==999.
    x = i.reshape(-1)[~mask.reshape(-1)]
    y = j.reshape(-1)[~mask.reshape(-1)]

    points = np.stack([x, y]).T
    values = array.reshape(-1)[~mask.reshape(-1)]
    print(values)
    # assert points.shape == values.shape, f"{points.shape}  {values.shape}"
    grid_z1 = griddata(points, values, (i.reshape(-1)[mask.reshape(-1)], 
                                        j.reshape(-1)[mask.reshape(-1)]), 
                                       method='nearest')
    z = np.copy(array)
    z[i.reshape(-1)[mask.reshape(-1)], j.reshape(-1)[mask.reshape(-1)]] = grid_z1
    
    return z

# plt.figure()
# out = interp(0)
# np.savetxt('RR_2022\inputs\lay_1_bot.csv', X = out, delimiter = ',', fmt = '%.1f')
# plt.imshow(out)

# plt.figure()
# out = interp(1)
# np.savetxt('RR_2022\inputs\lay_2_bot.csv', X = out, delimiter = ',', fmt = '%.1f')
# plt.imshow(out)

# plt.figure()
# out = interp(2)
# np.savetxt('RR_2022\inputs\lay_3_bot.csv', X = out, delimiter = ',', fmt = '%.1f')
# plt.imshow(out)

# c = "C:\modeling\MirabelWohler\RR_2022\mult\zones_hk.txt"
# array = np.genfromtxt(c, dtype = int)

# out = interp(0, array)
# plt.figure()
# plt.imshow(out)
# np.savetxt(c, out, fmt = '%d')

# c = "C:\modeling\MirabelWohler\RR_2022\mult\zones_ss.txt"
# array = np.genfromtxt(c, dtype = int)

# out = interp(0, array)
# plt.figure()
# plt.imshow(out)
# np.savetxt(c, out, fmt = '%d')


# c = "C:\modeling\MirabelWohler\RR_2022\mult\zones_sy.txt"
# array = np.genfromtxt(c, dtype = int)

# out = interp(0, array)
# plt.figure()
# plt.imshow(out)
# np.savetxt(c, out, fmt = '%d')

for file in ['hk1', 'hk2', 'ss1', 'ss2', 'sy1', 'sy2', 'ss3']:
    c = f"C:\modeling\MirabelWohler\RR_2022\pilot_points\{file}.txt"
    # "C:\modeling\MirabelWohler\RR_2022\pilot_points\hk1.txt"
    array = np.genfromtxt(c)
    lay = int(file[-1])
    print(lay)
    out = interp(0, array)
    plt.figure()
    plt.imshow(out)
    np.savetxt(c, out)
```
